# MOVIMIENTO BROWNIANO DRIFT

## Módulos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-colorblind') #alternative
plt.rcParams['figure.dpi'] = 300
%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

## Definición de Parámetros

```python
# Número de subintervalos
N = 10000
# Horizonte de tiempo (hasta donde vamos a graficar las trayectorias)
T = 1.  
# Tamaño del incremento
dt = T/N
# Replicabilidad de resultados
np.random.seed(100)
```

## Una Trayectoria (simulación)

```python
# Número de simulaciones
n_sims = 1

# Intervalo de tiempo
tiempos = np.linspace(0, T, N)
```

```python
# MBE

dB = np.random.normal(scale= np.sqrt(dt), size=(n_sims, N-1))
B = np.cumsum(dB, axis=1)
mbe = np.insert(B, 0, 0.0, axis=1)
```

```python
# Movimiento Browniano Drift

# Media
mu = 0.5
# Desviación
sigma = 0.4

# MBD
mbd = mu*tiempos + sigma*mbe
```

```python
# Gráfica
fig, ax = plt.subplots(1,1, figsize=(10,8), dpi=70)
ax.set_title('Movimiento Browniano Drift', fontsize=16)
ax.plot(tiempos, np.transpose(mbd))
# Línea de tendencia
ax.plot(tiempos, mu*tiempos)
ax.grid(True)
```

## Distintos valores de mu

```python
# Lista de valores de N (# subintervalos)
mu = [-1.2, -0.6, 0.4, 1.8]

# Desviación
sigma = 0.6
```

```python
# Gráfica
fig, ax = plt.subplots(4, figsize=(10,8))
ax[0].set_title('Movimiento Browniano Drift', fontsize=16)

for i in range(0,4):
  #tiempos = np.linspace(0, T, vN[i])
  #dt=T/vN[i]
  #dB = np.random.normal(scale= np.sqrt(dt), size=(n_sims, vN[i]-1))
  #B = np.cumsum(dB, axis=1)
  #mbe = np.insert(B, 0, 0.0, axis=1)
  mbd = mu[i]*tiempos + sigma*mbe
  ax[i].plot(tiempos, np.transpose(mbd),label=f'mu={mu[i]:.2f}')
  ax[i].plot(tiempos, mu[i]*tiempos, 'r')
  ax[i].legend(loc='upper left');
  ax[i].grid(True)
```
